##### Using zip keyword

```python
names = ['Tom','Tim','Jon','Jim','Kim']
marks = [89,90,97,87]
```

```python
zip(names,marks)
```

```python
for x in zip(names,marks):
    print(x)
```

```python
dict(  zip(names,marks)     )
```

```python
states = ['Karnataka','MP','UP','TN','AP']
capitals = ['Bengaluru','Bhopal','Lucknow','Chennai','Amaravathi']
```

```python
dstates = dict(zip(states,capitals))
dstates
```
