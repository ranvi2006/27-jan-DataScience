### Jupyter Shortcuts
+ esc + a - Add a cell above the current
+ esc + b - Add a cell below the current cell
+ esc + m - Convert current cell to markdown
+ esc + y - Convert a cell to code call
+ esc + x - Delete the current cell

### Iterating Over sequences

```python
shop_lst = ['Apples','Orange','Rice','Butter','Soap','Shampoo']
shop_lst
```

```python
for elem in shop_lst:
    print(elem)
    print('Hi')
```

```python
for elem in shop_lst:
    print(elem)
print('Hi')
```

```python
s1 = 'python3.0'

for x in s1:
    print(x,' -> ',type(x))
```

```python
for i in range(9):
    print(s1[i])
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1,11, 2):
    print(i)
```
