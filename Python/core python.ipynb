#### For loops

```python
marks_lst = [67,89,43,23,76,54]
```

```python
for mark in marks_lst:
    if mark >= 80:
        res = 'Distinction'
    elif mark >= 60:
        res = "Fist Class"
    elif mark >= 40:
        res = 'Second Class'
    else:
        res = 'Failed'
    print(mark, ' -> ', res)
```

#### Functions

```python
def avg(n1,n2):
    total = n1 + n2
    avg = total / 2
    return avg
```

```python
res = avg(10,20)
print('Result=',res)
```

```python
def avg2(n1,n2):
    total = n1 + n2
    avg = total / 2
    print(avg)
```

```python
res = avg2(10,20)
print('Result=',res)
```

```python
def find_grade(mark):
    if mark >= 80:
        res = 'Distinction'
    elif mark >= 60 and mark < 80:
        res = "Fist Class"
    elif mark >= 40 and mark < 60:
        res = 'Second Class'
    else:
        res = 'Failed'
    return res
```

```python
grade = find_grade(98)
grade
```

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact * i
    return fact
```

```python
factorial(6)
```
